```python
import pandas as pd
df = pd.read_csv("./data/uber_eats_orders.csv")

df.head()
```

```python
df.info()
df.describe()
```

```python
df["is_peak_hour"] = df["order_hour"].between(11, 13) | df["order_hour"].between(18, 21)
df["price_after_surge"] = df["order_value"] * df["surge_multiplier"]
df["is_peak_hour"] 
df["price_after_surge"]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


plt.figure()
df["delivery_time_min"].hist(bins=30)
plt.title("Delivery Time Distribution")
plt.xlabel("Minutes")
plt.ylabel("Frequency")
plt.show()
```

```python
#Cell 5  Bivariate Analysis

plt.figure()
sns.scatterplot(
    data=df,
    x="distance_km",
    y="delivery_time_min",
    alpha=0.4
)
plt.title("Delivery Time vs Distance")
plt.show()
```

```python
# Cell 6  Ratings vs Delivery Time

plt.figure()
sns.boxplot(
    data=df,
    x="traffic_level",
    y="delivery_time_min"
)
plt.title("Delivery Time by Traffic Level")
plt.show()

```

```python
# Cuisine Performance

cuisine_stats = df.groupby("cuisine")[["rating", "delivery_time_min"]].mean().sort_values("rating", ascending=False)
cuisine_stats
```

```python
# Correlation Heatmap

plt.figure()
sns.heatmap(
    df[["distance_km", "prep_time_min", "delivery_time_min", "order_value", "rating"]].corr(),
    annot=True,
    cmap="coolwarm"
)
plt.title("Correlation Matrix")
plt.show()
```
